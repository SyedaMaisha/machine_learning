```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

model = LogisticRegression()

model.fit(x_train_scaled, y_train) 
y_pred = model.predict(x_test_scaled)
print(y_pred)
```

```python
accuracy1=accuracy_score(y_test, y_pred)
print( "Accuracy: " ,accuracy1)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(criterion='entropy',random_state=1)
clf.fit(x_train_scaled,y_train)
y_pred = clf.predict(x_test_scaled)
accuracy2=accuracy_score(y_pred,y_test)
print( "Accuracy: " ,accuracy2)
```

```python
graph= plt.figure(figsize=(5,5))
addressed= ['Logistic regression','Decision tree']
accuracy=[accuracy1*100, accuracy2*100]
position=[0,1]
plt.bar(position,accuracy,width=0.5)
plt.xticks(position,addressed)
plt.title('Accuracy visualization')
plt.ylabel("Accuracy %")
plt.show()
```
